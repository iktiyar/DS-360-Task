## Question 1: Basic Measures of Central Tendency and Dispersion (Beginner)
**Points: 10**  

A retail store manager collected data on the number of customers per day for the past week:
Data: 45, 52, 38, 47, 55, 41, 49  

**Part A (4 points):** Calculate the mean, median, and mode for this dataset.  

**Part B (3 points):** Calculate the sample variance and sample standard deviation. Show your work
step by step.  

**Part C (3 points):** Explain which measure of central tendency (mean, median, or mode) would be
most appropriate if there was an outlier day with 150 customers, and why?

## Answer 1
**Part A:**  

Sorting given data in ascending order - 38, 41, 45, 47, 49, 52, 55  

Here n = 7, an odd number.  

mean = (38+41+....+55) / 7 = **46.71**  

median = (n+1)/2 th number = 4th number = **47**  

mode = **None**  
  
**Part B:**  
Let, 

x = [45, 52, 38, 47, 55, 41, 49]  

x - $\bar{x}$ = [-1.71,  5.29, -8.71,  0.29,  8.29, -5.71,  2.29]  

(x - $\bar{x})^2$ = [ 2.94, 27.94, 75.94,  0.08, 68.65, 32.65,  5.22] 

$\Sigma(x - \bar{x})^2$ = 213.42  

sample variance, $s^2$ = 213.42 / (n-1) = 213.42/6 = 35.57  

sample standard deviation, s = $\sqrt{s^2} = \sqrt{35.57}$ = 5.96  

**Part C:**  

Data - 38, 41, 45, 47, 49, 52, 55, 150  

mean = (38+41+....+55+150) / 8 = **59.62**  

median = Mean of (n/2)th and (n/2)+1 th number = Mean of (4th and 5th number) = **48**  

mode = **None**

**The Median would be the most approprate if an outlier (150) appears**. From part A, we can see mean is 46.71 but in part C mean increases to  59.62 which is large deviation from part A. On the other hand median changes slightly from 47 to 48. Thus, in this case median is a more robust statistic compared to mean in the presence of outliers.

## Question 2: Population vs Sample and Variance Division (Beginner-Intermediate)  

**Points: 12**  

A quality control manager wants to analyze the weight of chocolate bars produced in a factory. The
population of all chocolate bars has a mean weight of μ = 100g and standard deviation σ = 5g.  

**Part A (4 points):** Explain the difference between population and sample in this context. Why might
the manager choose to work with a sample rather than the entire population?  

**Part B (5 points):** The manager takes a sample of 8 chocolate bars with weights: 98, 102, 95, 105,
99, 103, 97, 101g. Calculate the sample variance using the formula with (n-1) in the denominator.  

**Part C (3 points):** Explain why we divide by (n-1) instead of n when calculating sample variance.
What statistical concept does this address?

## Answer 2  

**Part A:**  
Population: The popuation is all the produced chocolate bars in the factory which has a mean weight, $\mu = 100g$ and standard deviation, $\sigma = 5g$.  

Sample: A subset from a population is called a sample. For example, some chocolate bars are randomly taken from the factory.

- It is costly and time consuming to work with the entire population.
- Can get a good characteristics overview of a population by working with a good and reresentative sample.  

**Part B:**  

Let, 

n= 8 

x =  [98, 102, 95, 105, 99, 103, 97, 101] 

x - $\bar{x}$ =  [-2,  2, -5,  5, -1,  3, -3,  1]  

(x - $\bar{x})^2$ =  [4,  4, 25, 25,  1,  9,  9,  1]

$\Sigma(x - \bar{x})^2$ = 78

sample variance, $s^2$ = 78 / (n-1) = 78/7 = 11.14 

sample standard deviation, s = $\sqrt{s^2} = \sqrt{11.14}$ = 3.34g  

**Part C:**  

When calculating sample variance, we divide by (n-1) instead of n because this adjustment increases the sample variance slightly and reduce the biases that occurs when estimating population mean using sample mean. Also, sample mean itself an estimate of popuation mean, thus we lose one degree of freedom. This adjustment is known as Bessel's correction.

## Question 4: Distribution Types and Skewness (Intermediate)  

**Points: 18**

An online learning platform analyzed the completion times (in minutes) for a statistics course. They found three different patterns in their data:  
Dataset A: Mean = 120 min, Median = 115 min, Mode = 110 min  
Dataset B: Mean = 85 min, Median = 90 min, Mode = 95 min  
Dataset C: Mean = 95 min, Median = 95 min, Mode = 95 min  

**Part A (9 points):** For each dataset, determine the type of skewness (left skewed, right skewed, or
symmetrical) and explain your reasoning based on the relationship between mean, median, and mode.  

**Part B (6 points):** Give two real-world examples of data that would naturally follow each type of
distribution (left skewed, right skewed, symmetrical) and explain why.  

**Part C (3 points):** Which measure of central tendency would be most reliable for making business
decisions in each of the three datasets? Justify your answer.

## Answer 4  

**Part A:**  

Dataset A: Mean > Median > Mode **(Right-Skewed or Positively Skewed, Mean towards the higher values creating a right tail.)**  
Dataset B: Mean < Median < Mode **(Left-Skewed or Negatively Skewed, Mean towards the smaller values creating a left tail)**  
Dataset C: Mean = Median = Mode **(Symmetrical or No Skew, Values are symmetrically distributed around center.)**  

**Part B:**  

Right-skewed:
1. Time spent completing online courses
2. Income distribution in a city  
- Most people complete courses in a moderate time, but a few take very long,long right tail. Similarly, most earn average income, but few earn extremely high income. 

Left-skewed
1. Age at retirement
2. Test scores in an easy exam  
- Most retire at a typical age, but some retire early,long left tail. In easy exams, most students score high, but a few score very low, left tail.

Symmetrical
1. Heights of adult men in a population
2. IQ scores  
- Data is evenly distributed around the mean, forming a bell-shaped pattern.

**Part C:**  

For dataset A and B, median would be the most reliable for making business decisions. Since, the distribution of these two datasets are skewed, mean is affected by extremely large or small values in the tail of the distribution, whereas median is more robust and not influenced by extreme values.  
Dataset C has a symmetrical distribution. Therefore, mean is reliable measure here because no extreme values affecting the central tendancy.

## Question 5: Z-Scores and Standardization(Intermediate)  

**Points: 13**  

A university professor finds that exam scores follow a normal distribution with μ = 78 and σ = 12.  

**Part A (8 points):** Calculate the z-scores for the following students:  
Student A: 90 points  
Student B: 65 points  
Student C: 78 points  
Student D: 95 points  
Show your calculations and interpret what each z-score means.  

**Part C (5 points):** If the professor wants to convert all scores to a standard normal distribution (mean
= 0, standard deviation = 1), what would be the new standardized scores for Students A and B?
Explain the advantages of standardization in this context.

## Answer 5  

**Part A:**  

Given, $\mu = 78$ and $\sigma = 12$  

We know, $ z = \frac{X - \mu}{\sigma}$  

For student A (90 points), $z_A$ = (90 − 78) / 12 =  12 / 12 = 1.00 (Obtained 1 standard deviation above mean)  

For student B (65 points), $z_B$ = (65 − 78) / 12 =  −13 / 12 = −1.08  (Obtained 1.08 standard deviation below mean)  

For student C (78 points), $z_C$ = (78 − 78) / 12 =  0 / 12 = 0  (Obtained equal to mean)  

For student D (95 points), $z_D$ = (95 − 78) / 12 =  17 / 12 = 1.42 (Obtained 1.42 standard deviation above mean)  

**Part C:**  

For, $\mu = 0$ and $\sigma = 1.00$  

New, $z_A$ = 1.00 and $z_B$ = -1.08  

Advantages:  
- All scores are cmpacted to same scale.  
- Provide instant camparison with mean.
